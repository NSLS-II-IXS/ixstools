```python
from ixstools import align
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib notebook
import numpy as np
```

```python
def plot(xy):
    fig, ax = plt.subplots()
    arts = {}
    for sid, (xdata, ydata) in enumerate(xy):
        arts[sid] = ax.plot(xdata, ydata, '-o', label=sid)
    ax.legend(loc=0)
```
